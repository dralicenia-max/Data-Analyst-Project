```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

```python
complt = pd.read_csv(r'C:\Users\glori\OneDrive\Documents\Desktop\DCWP_Consumer_Complaints_20260425.csv')
```

```python
complt.head()
```

```python
complt['Borough'] = complt['Borough'].str.strip().str.lower()

manhattan_counts = (complt[complt['Borough'] == 'manhattan'].groupby('Complaint Code')['Borough'].count().rename('Manhattan').sort_values(ascending=False).head(10))


queens_counts = (complt[complt['Borough'] == 'queens'].groupby('Complaint Code')['Borough'].count().rename('Queens').sort_values(ascending=False).head(10))


compar_df = pd.concat([manhattan_counts, queens_counts], axis=1).fillna(0).astype(int)


print(compar_df)
```

```python
compar_df = compar_df.reset_index()


df_melted = compar_df.melt(
    id_vars='Complaint Code',
    value_vars=['Manhattan', 'Queens'],
    var_name='Borough',
    value_name='Count'
)

sns.set_theme(style="whitegrid")


plt.figure(figsize=(12, 6))
sns.barplot(
    data=df_melted,
    x='Complaint Code',
    y='Count',
    hue='Borough'
)


plt.xticks(rotation=45, ha='right')


plt.xlabel("Complaint Code")
plt.ylabel("Number of Complaints")
plt.title("Top 10 Complaint Codes in Manhattan vs Queens")


plt.legend(title="Borough")


plt.tight_layout()
plt.show()
```
