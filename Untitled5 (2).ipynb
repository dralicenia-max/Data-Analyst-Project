```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
nypd = pd.read_csv(r'C:\Users\glori\OneDrive\Documents\Desktop\NYPD_Arrest_Data_Year_to_Date.csv')
```

```python
#Compare two precinct 115 Jackson Height with the precint with more incidents 14
compare = nypd[nypd['ARREST_PRECINCT'].isin([115, 14])] \
    .groupby(['OFNS_DESC', 'ARREST_PRECINCT']) \
    .size() \
    .unstack(fill_value=0)
compare['total'] = compare.sum(axis=1)
compare = compare.sort_values(by='total', ascending=False)
print(compare)
```

```python
plot_df = compare.drop(columns=['total'])
plot_df.plot(kind='bar', figsize=(12,6))
plt.title('Arrests by Offense Type: Precinct 14 vs 115')
plt.xlabel('Offense Description')
plt.ylabel('Number of Arrests')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Precinct')
plt.tight_layout()
plt.show()
```
